```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,20)
```

```python
y1=2*x
y2=x+4
```

```python
y1
```

```python
y2
```

```python
x
```

```python
plt.plot(x,y1,color='r',linestyle='-.',linewidth=5)
plt.plot(x,y2,color='g',linestyle=':',linewidth=5)
plt.title('ST LINE')
plt.ylabel('price')
plt.xlabel('no of items')
plt.show()
```

```python
#adding subplot
x1=np.arange(1,20,2)
x2=np.arange(1,40,4)
y1=x1*x1
y2=x2*x2
```

```python
plt.subplot(1,2,1)#1st argument=norows,2nd argument=nocolumns
plt.plot(x1,y1,color='r',linestyle='-.',linewidth=5)
plt.title('ST LINE')
plt.ylabel('price')
plt.xlabel('no of items')

plt.subplot(1,2,2)
plt.plot(x2,y2,color='g',linestyle=':',linewidth=5)
plt.title('ST LINE')
plt.ylabel('price')
plt.xlabel('no of items')
plt.show()
```

```python
# bar graph
studentsdata={'ram':20,'shyam':30,'mohan':25}
```

```python
names=list(studentsdata.keys())
marks=list(studentsdata.values())
plt.bar(names,marks,color='g')#barh for horizontal graph
plt.title('performance')
plt.xlabel('names')
plt.ylabel('marks')
plt.grid(True)
plt.show()
```
